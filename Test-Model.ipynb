```python
import tensorflow as tf 
import cv2
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
CATEGORIES = ["elliptical","spiral"]
IMAGE_SIZE = 100
```

```python
image1 = cv2.imread('spiral-galaxy.jpg')
image1 = cv2.resize(image1,(IMAGE_SIZE,IMAGE_SIZE))
plt.imshow(image1)
```

```python
image2 = cv2.imread('ellipse-galaxy.jpg')
image2 = cv2.resize(image2,(IMAGE_SIZE,IMAGE_SIZE))
plt.imshow(image2)
```

```python
model = tf.keras.models.load_model('CNN-Classifier.h5')
```

```python
x = model.predict(image2.reshape(-1,IMAGE_SIZE,IMAGE_SIZE,3))
x = np.argmax(x)
CATEGORIES[x]
```
